El siguiente código es un ejemplo de cómo podríamos estimar una integral definida usando Montecarlo

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definición de la función a integrar e intervalo de integración

```python
# Límites de integración
a = 0
b = 2  

def funcion_a_integrar(x):
    return  x**2

N = 1000
```

```python
y = np.random.uniform(funcion_a_integrar(a), funcion_a_integrar(b), N)
x = np.random.uniform(a, b, N)
```

```python
n_bajo_curva = y < funcion_a_integrar(x)
```

```python
n_bajo_curva = np.sum(y < funcion_a_integrar(x))
n_bajo_curva / N
```

```python
respuesta_de_integrar = (b-a) * funcion_a_integrar(b) * n_bajo_curva / N
respuesta_de_integrar
```

```python
plt.plot(x, y, 'o')
plt.plot(np.linspace(a, b, 10), funcion_a_integrar(np.linspace(a, b, 10)))
plt.show()
```

```python
""""
n_simulaciones: numero de simulaciones de montecarlo
L : largo de las agujas a soltar
d : Distancia entre las lineas
""" 
```

```python
def estimar_pi_buffon(n_simulaciones = 100000, L=1, d=2):
    cruces = 0

    for _ in range(n_simulaciones):
        #simulas dond cae el centro en una altura aleatoria dentro de una linea 
        # aleatorio uniforme continuo entre 0 y d/2
        centro_de_aguja = np.random.uniform(0, d/2)

        #simulas el angulo con el que cae la aguja 
        theta = np.random.uniform(0, np.pi/2)

        #la aguja cruzó?
        if (L/2) * np.sin(theta) >= centro_de_aguja:
            cruces += 1 

    # calculas pi
    # formas alternativas: (2 * numero de lineas * largo de aguja) / (distancia entre lineas * agujas que cruzan)
    pi_estimado = 2 * (n_simulaciones * L) / (d * cruces)

    return pi_estimado

pi_estimado = estimar_pi_buffon()
print(pi_estimado)
```

```python

import matplotlib.pyplot as plt

def estimar_pi(n_puntos=10000):
    x = np.random.rand(n_puntos)  # entre 0 y 1
    y = np.random.rand(n_puntos)  # entre 0 y 1
    
    # puntos dentro del cuarto de círculo
    dentro = x**2 + y**2 <= 1
    
    pi_estimado = 4 * np.sum(dentro) / n_puntos
    
    # gráfica 
    plt.scatter(x[dentro], y[dentro], s=1, color="blue")
    plt.scatter(x[~dentro], y[~dentro], s=1, color="red")
    plt.gca().set_aspect("equal")
    plt.title(f"π ≈ {pi_estimado:.5f}")
    plt.show()
    
    return pi_estimado

print("π estimado:", estimar_pi(10000))
```
